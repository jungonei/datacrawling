```python
import pandas as pd
from bs4  import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.tokopedia.com/search?st=product&q=sepatu')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
rows = soup.find_all('div', class_='css-974ipl')
```

```python
data = []
for row in rows :
    Nama = row.find(class_='prd_link-product-name css-3um8ox').text.strip()
    Harga = row.find(class_='prd_link-product-price css-1ksb19c').text.strip()    
    Rating = row.find(class_='prd_rating-average-text css-t70v7i').text.strip()
    Terjual = row.find(class_='prd_label-integrity css-1duhs3e').text.strip()
    
    rich = {
        'Nama': Nama, 
        'Harga': Harga,
        'Rating': Rating,
        'Terjual': Terjual,
    }
    
    data.append(rich)
```

```python
pd.DataFrame(data).to_excel('Tokopedia_3.xlsx')
```
